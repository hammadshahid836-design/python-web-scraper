#**What is Web Scraping**
- Web scraping is the automated process of extracting data from websites by sending requests to:
- Web pages,
- Retrieving their HTML/JSON/XML content, and
- then parsing and converting the required information into a structured format (like CSV, Excel, database, or DataFrame) for analysis or reuse.

```
import requests
from bs4 import BeautifulSoup

base_url = "https://books.toscrape.com/catalogue/page-{}.html"

# Loop through pages 1 to 50
for page in range(1, 51):
    url = base_url.format(page)
    resp = requests.get(url, timeout=30)

    if resp.status_code == 200:
        soup = BeautifulSoup(resp.text, "html.parser")
        containers = soup.find_all("h3")

        print(f"\nPage {page}: Found {len(containers)} books")

        # Extract book titles
        for h3 in containers:
            title = h3.find("a")["title"]   # book title is inside the "title" attribute
            print(" -", title)
    else:
        print(f"Page {page}: Failed to fetch (status {resp.status_code})")
```

# Step 4:
Locate elements using tags, classes, or CSS selectors

```
# STEP 4: Select containers for each item
cards = soup.select("article.product_pod")
```

```
rows = []
rating_map = {"One":1, "Two":2, "Three":3, "Four":4, "Five":5}
```

# Step 5:
Repeat process for all required data

```
import re
# Regular expression, remove symbols, extract digits, filter URLs, etc
rows = []
for card in cards:
    title = card.h3.a.get("title", "").strip()

    # --- clean price ---
    price_text = card.select_one(".price_color").get_text(strip=True)
    clean_price = re.sub(r"[^\d.]", "", price_text)   # removes £, Â, etc.
    price = float(clean_price) if clean_price else None

    # rating
    rating_word = card.select_one("p.star-rating")["class"][1]
    rating = rating_map.get(rating_word, None)

    # availability
    availability = card.select_one(".instock.availability").get_text(strip=True)

    # append row (no URL)
    rows.append({
        "title": title,
        "price_gbp": price,
        "rating": rating,
        "availability": availability
    })
```

# Step 6:
Save your file in required format

```
# STEP 5: Convert to DataFrame
df = pd.DataFrame(rows)
print(df.head())

# STEP 6: Save output
df.to_csv("books_page1.csv", index=False)
```

# Step 7:
Repeat above steps for many pages or use loop

```
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

# Base URL of the website
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
book_base_url = 'https://books.toscrape.com/catalogue/'

# Create directories if they don't exist
os.makedirs('data_sheet', exist_ok=True)
os.makedirs('images', exist_ok=True)

# List to store book details
books = []

# Function to extract data from a single page
def extract_data_from_page(soup, page):
    for book in soup.find_all('article', class_='product_pod'):
        title = book.h3.a['title']
        price = book.find('p', class_='price_color').text
        availability = book.find('p', class_='instock availability').text.strip()
        rating_text = book.p['class'][1]
        rating = convert_rating_to_number(rating_text)
        link = book_base_url + book.h3.a['href']
        thumbnail_url = 'https://books.toscrape.com/' + book.find('img', class_='thumbnail')['src']
        thumbnail_file_name = save_thumbnail_image(thumbnail_url, title)

        book_data = {
            'Title': title,
            'Price': price,
            'Availability': availability,
            'Rating': rating,
            'Link': link,
            'Thumbnail URL': thumbnail_url,
            'Thumbnail File Name': thumbnail_file_name
        }

        books.append(book_data)

        # Print the extracted data to the console in a formatted way
        print(f"Page: {page}")
        print(f"Title: {title}")
        print(f"Price: {price}")
        print(f"Availability: {availability}")
        print(f"Rating: {rating}")
        print(f"Link: {link}")
        print(f"Thumbnail URL: {thumbnail_url}")
        print(f"Thumbnail File Name: {thumbnail_file_name}")
        print("="*50)

# Function to convert rating text to a number
def convert_rating_to_number(rating_text):
    rating_dict = {
        'One': 1,
        'Two': 2,
        'Three': 3,
        'Four': 4,
        'Five': 5
    }
    return rating_dict.get(rating_text, 0)

# Function to save thumbnail image
def save_thumbnail_image(url, title):
    response = requests.get(url)
    sanitized_title = "".join([c if c.isalnum() else "_" for c in title])
    image_path = os.path.join('images', f"{sanitized_title}.jpg")
    with open(image_path, 'wb') as file:
        file.write(response.content)
    return image_path

# Loop through the first 10 pages
for page in range(1, 51):  # Scraping the first 10 pages
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    extract_data_from_page(soup, page)
    print(f"Extracted data from page {page}")

# Save the data to a CSV file
df = pd.DataFrame(books)
df.to_csv('data_sheet/books_data.csv', index=False)
print("Data saved to data_sheet/books_data.csv")
```
